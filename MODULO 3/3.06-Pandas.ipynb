#### Q4. 
* Crie um dataframe `gols` com duas colunas `pa√≠s` e `total_gols` com o total de gols marcados por cada pa√≠s durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
pa√≠s        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
import pandas as pd

# Supondo que wc seja o DataFrame original

# 1. Somar os gols de cada pa√≠s
gols_casa = wc.groupby('time_1')['gols_1'].sum().reset_index()
gols_visitante = wc.groupby('time_2')['gols_2'].sum().reset_index()

# 2. Juntar os gols de casa e visitante
gols_casa.columns = ['pa√≠s', 'gols_casa']
gols_visitante.columns = ['pa√≠s', 'gols_visitante']

# 3. Combinar os DataFrames e calcular o total de gols
gols = pd.merge(gols_casa, gols_visitante, on='pa√≠s', how='outer').fillna(0)
gols['total_gols'] = gols['gols_casa'] + gols['gols_visitante']

# 4. Selecionar apenas as colunas 'pa√≠s' e 'total_gols'
gols = gols[['pa√≠s', 'total_gols']]

# 5. Ordenar o DataFrame pelo total de gols em ordem decrescente
gols = gols.sort_values(by='total_gols', ascending=False)

# Exibir o DataFrame resultante
print(gols)
````

#### Q5. 
Qual pa√≠s tomou mais cart√µes amarelos somando todas as copas?

Neste exerc√≠cio voc√™ vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observa√ß√£o √© uma string represetando uma lista dos cart√µes amarelos de um √∫nico jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16‚Äô|1:0|Rosana G√≥mez', '20‚Äô|2:0|Gabriela Ch√°vez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o m√©todo gen√©rico `apply` para chamar uma fun√ß√£o que voc√™ vai criar para transformar uma observa√ß√£o de cart√£o amarelo em uma contagem de elementos da lista. 

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Em seguida fa√ßa sua m√°gica para agrupar as informa√ß√µes por pa√≠s, acumular os cart√µes de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
pa√≠s        cart√µes amarelos
Argentina   97
England     93
Australia   93
...
```

````python
import pandas as pd
import ast  # Para converter strings em listas

# Supondo que wc seja o DataFrame original

# 1. Fun√ß√£o para contar cart√µes amarelos
def contar_cartoes(observacao):
    if pd.isna(observacao):  # Se a observa√ß√£o for NaN, retornar 0
        return 0
    try:
        lista_cartoes = ast.literal_eval(observacao)  # Converter string em lista
        return len(lista_cartoes)  # Retornar o n√∫mero de cart√µes
    except:
        return 0  # Caso ocorra algum erro, retornar 0

# 2. Aplicar a fun√ß√£o √†s colunas cartao_amarelo_1 e cartao_amarelo_2
wc['num_cartoes_1'] = wc['cartao_amarelo_1'].apply(contar_cartoes)
wc['num_cartoes_2'] = wc['cartao_amarelo_2'].apply(contar_cartoes)

# 3. Agrupar os dados por pa√≠s e somar os cart√µes amarelos
cartoes_casa = wc.groupby('time_1')['num_cartoes_1'].sum().reset_index()
cartoes_visitante = wc.groupby('time_2')['num_cartoes_2'].sum().reset_index()

# Renomear as colunas
cartoes_casa.columns = ['pa√≠s', 'cartoes_casa']
cartoes_visitante.columns = ['pa√≠s', 'cartoes_visitante']

# Combinar os DataFrames
cartoes_total = pd.merge(cartoes_casa, cartoes_visitante, on='pa√≠s', how='outer').fillna(0)
cartoes_total['cart√µes amarelos'] = cartoes_total['cartoes_casa'] + cartoes_total['cartoes_visitante']

# Selecionar apenas as colunas 'pa√≠s' e 'cart√µes amarelos'
cartoes_total = cartoes_total[['pa√≠s', 'cart√µes amarelos']]

# 4. Ordenar o DataFrame pelo total de cart√µes amarelos em ordem decrescente
cartoes_total = cartoes_total.sort_values(by='cart√µes amarelos', ascending=False)

# Exibir o resultado
print(cartoes_total)
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de p√™nalti.

Para conseguir essa informa√ß√£o, voc√™ precisar√° trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas tamb√©m s√£o colunas textuais, onde cada observa√ß√µes apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol est√° na forma `'Jogador(a) ¬∑ minuto‚Äô'`. Por exemplo:
```
'Alex Morgan ¬∑ 12‚Äô|Rose Lavelle ¬∑ 20‚Äô'
```

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Recomendo criar um dicion√°rio √† parte, onde cada chave ser√° um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente ser√° a contagem de ocorr√™ncias desses nomes.

Em seguida basta converter o seu dicion√°rio em um dataframe e imprim√≠-lo na forma (valores n√£o s√£o inventados üòÅ):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
import pandas as pd

# Supondo que wc seja o DataFrame original

# 1. Fun√ß√£o para extrair os jogadores que marcaram gols
def extrair_jogadores(observacao):
    if pd.isna(observacao):  # Se a observa√ß√£o for NaN, retornar lista vazia
        return []
    try:
        # Dividir a string pelos caracteres '|' e extrair o nome do jogador
        gols = observacao.split('|')
        jogadores = [gol.split('¬∑')[0].strip() for gol in gols]  # Extrair o nome do jogador
        return jogadores
    except:
        return []  # Caso ocorra algum erro, retornar lista vazia

# 2. Processar as colunas de gols
colunas_gols = ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']
dicionario_gols = {}  # Dicion√°rio para armazenar a contagem de gols

for coluna in colunas_gols:
    for observacao in wc[coluna]:
        jogadores = extrair_jogadores(observacao)
        for jogador in jogadores:
            if jogador in dicionario_gols:
                dicionario_gols[jogador] += 1
            else:
                dicionario_gols[jogador] = 1

# 3. Converter o dicion√°rio em um DataFrame
df_gols = pd.DataFrame(list(dicionario_gols.items()), columns=['jogador(a)', 'num_gols'])

# 4. Ordenar o DataFrame pelo n√∫mero de gols em ordem decrescente
df_gols = df_gols.sort_values(by='num_gols', ascending=False)

# 5. Exibir o top 10
print(df_gols.head(10))
````
