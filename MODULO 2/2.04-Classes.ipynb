#### Q1.
Vamos criar aos pouquinhos uma classe Evento que registra eventos de um calend√°rio. Pra come√ßar, nessa quest√£o voc√™ deve:

* Criar uma classe chamada Evento
* Adicionar um atributo de classe `total_eventos` que ser√° usado para contar o n√∫mero total de eventos (inst√¢ncias de classe) criadas.
* Implemente o m√©todo construtor que deve receber os par√¢metros `titulo` (`string`), `data_hora` (`datetime`), `descri√ß√£o` (`string`). Crie e inicialize um atributo de inst√¢ncia para cada um dos par√¢metros citados.
* O construtor tamb√©m deve inicializar um atributo de inst√¢ncia `is_concluido = False`. Esse atributo ser√° utilizado mais √† frente. 
* No construtor voc√™ tamb√©m deve incrementar o atributo de classe `total_eventos` em 1 a cada nova inst√¢ncia criada.

Teste o seu c√≥digo criando duas diferentes inst√¢ncias de evento (preencha os atributos como quiser) e imprimindo todos os seus atributos (de classe e de inst√¢ncia).

> Consulte [este tutorial biblioteca `datetime`](https://www.w3schools.com/python/python_datetime.asp) caso n√£o esteja familiarizado.

```python
from datetime import datetime

class Evento:
    # Atributo de classe para contar o n√∫mero total de eventos
    total_eventos = 0

    def __init__(self, titulo, data_hora, descricao):
        # Atributos de inst√¢ncia
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False  # Inicializado como False

        # Incrementa o contador de eventos
        Evento.total_eventos += 1

    def __str__(self):
        # M√©todo para representar o objeto como string
        return (f"T√≠tulo: {self.titulo}, "
                f"Data e Hora: {self.data_hora}, "
                f"Descri√ß√£o: {self.descricao}, "
                f"Conclu√≠do: {self.is_concluido}")

# Exemplo de uso
# Criando dois eventos
evento1 = Evento(
    titulo="Reuni√£o de Planejamento",
    data_hora=datetime(2023, 10, 15, 14, 30),
    descricao="Reuni√£o para planejar o pr√≥ximo trimestre."
)

evento2 = Evento(
    titulo="Anivers√°rio do Jo√£o",
    data_hora=datetime(2023, 10, 20, 18, 0),
    descricao="Festa de anivers√°rio do Jo√£o."
)

# Imprimindo os atributos das inst√¢ncias
print("Evento 1:")
print(evento1)
print("\nEvento 2:")
print(evento2)

# Imprimindo o atributo de classe total_eventos
print(f"\nTotal de eventos criados: {Evento.total_eventos}")
```

* Um m√©todo de classe `num_eventos()` que retorna o valor do atributo de classe `total_eventos`.

Para o teste, use a fun√ß√£o nativa do Python [`isinstance`](https://www.w3schools.com/python/ref_func_isinstance.asp).

```python
from datetime import datetime

class Evento:
    # Atributo de classe para contar o n√∫mero total de eventos
    total_eventos = 0

    def __init__(self, titulo, data_hora, descricao):
        # Atributos de inst√¢ncia
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False  # Inicializado como False

        # Incrementa o contador de eventos
        Evento.total_eventos += 1

    def isConcluido(self):
        """
        Verifica se a data e hora do evento j√° passaram.
        Se sim, atualiza o atributo is_concluido para True.
        """
        if self.data_hora < datetime.now():
            self.is_concluido = True
        return self.is_concluido

    @classmethod
    def num_eventos(cls):
        """
        M√©todo de classe que retorna o n√∫mero total de eventos criados.
        """
        return cls.total_eventos

    @staticmethod
    def valida_evento(titulo, data_hora, descricao):
        """
        M√©todo est√°tico que valida os tipos dos atributos de um evento.
        Retorna True se todos os tipos estiverem corretos, False caso contr√°rio.
        """
        return (isinstance(titulo, str) and
                isinstance(data_hora, datetime) and
                isinstance(descricao, str))

    def __str__(self):
        # M√©todo para representar o objeto como string
        return (f"T√≠tulo: {self.titulo}, "
                f"Data e Hora: {self.data_hora}, "
                f"Descri√ß√£o: {self.descricao}, "
                f"Conclu√≠do: {self.is_concluido}")

# Testando a classe atualizada

# 1. Criando uma inst√¢ncia de Evento e chamando o m√©todo isConcluido()
evento1 = Evento(
    titulo="Reuni√£o de Planejamento",
    data_hora=datetime(2023, 10, 15, 14, 30),  # Data no passado
    descricao="Reuni√£o para planejar o pr√≥ximo trimestre."
)

# Verificando se o evento est√° conclu√≠do
evento1.isConcluido()
print("Evento 1:")
print(evento1)
print(f"Est√° conclu√≠do? {evento1.is_concluido}\n")

# 2. Invocando o m√©todo de classe num_eventos()
print(f"Total de eventos criados: {Evento.num_eventos()}\n")

# 3. Chamando o m√©todo est√°tico valida_evento()
# Testando com valores corretos
valido = Evento.valida_evento(
    titulo="Festa de Anivers√°rio",
    data_hora=datetime(2023, 12, 25, 20, 0),
    descricao="Festa de anivers√°rio do Jo√£o."
)
print(f"Valida√ß√£o com valores corretos: {valido}")

# Testando com valores incorretos
invalido = Evento.valida_evento(
    titulo=123,  # T√≠tulo n√£o √© string
    data_hora="2023-12-25",  # Data n√£o √© datetime
    descricao=456  # Descri√ß√£o n√£o √© string
)
print(f"Valida√ß√£o com valores incorretos: {invalido}")
```

#### Q3.

Vamos incluir m√©todos m√°gicos! ‚ú®ü™ÑüîÆ

Redefina a classe incluindo":
* M√©todo `__str__` que imprime os atributos do evento na forma `"Evento: titulo, Data: data_hora, Descri√ß√£o: descricao, Concluido: is_concluido"`.
* Implemente os m√©todos de compara√ß√£o `__eq__`, `__ne__`, `__lt__`, `__le__`, `__gt__` e `__ge__` para comparar eventos baseados no atributo `data_hora`. Esses m√©todos devem comparar duas inst√¢ncias de Evento e retornar os resultados apropriados (`True` ou `False`).

Para testar, crie duas inst√¢ncias de Evento com datas diferentes. Imprima as inst√¢ncias com a fun√ß√£o `print()` e apresente o resultado das compara√ß√µes entre eventos (`==`, `!=`, `<`, `<=`, `>`, `>=`).

```python
from datetime import datetime

class Evento:
    # Atributo de classe para contar o n√∫mero total de eventos
    total_eventos = 0

    def __init__(self, titulo, data_hora, descricao):
        # Atributos de inst√¢ncia
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False  # Inicializado como False

        # Incrementa o contador de eventos
        Evento.total_eventos += 1

    def isConcluido(self):
        """
        Verifica se a data e hora do evento j√° passaram.
        Se sim, atualiza o atributo is_concluido para True.
        """
        if self.data_hora < datetime.now():
            self.is_concluido = True
        return self.is_concluido

    @classmethod
    def num_eventos(cls):
        """
        M√©todo de classe que retorna o n√∫mero total de eventos criados.
        """
        return cls.total_eventos

    @staticmethod
    def valida_evento(titulo, data_hora, descricao):
        """
        M√©todo est√°tico que valida os tipos dos atributos de um evento.
        Retorna True se todos os tipos estiverem corretos, False caso contr√°rio.
        """
        return (isinstance(titulo, str) and
                isinstance(data_hora, datetime) and
                isinstance(descricao, str))

    def __str__(self):
        """
        Retorna uma representa√ß√£o leg√≠vel do evento.
        """
        return (f"Evento: {self.titulo}, "
                f"Data: {self.data_hora}, "
                f"Descri√ß√£o: {self.descricao}, "
                f"Conclu√≠do: {self.is_concluido}")

    # M√©todos de compara√ß√£o
    def __eq__(self, outro):
        """Compara se dois eventos t√™m a mesma data e hora."""
        return self.data_hora == outro.data_hora

    def __ne__(self, outro):
        """Compara se dois eventos t√™m datas e horas diferentes."""
        return self.data_hora != outro.data_hora

    def __lt__(self, outro):
        """Compara se a data e hora deste evento √© anterior √† do outro."""
        return self.data_hora < outro.data_hora

    def __le__(self, outro):
        """Compara se a data e hora deste evento √© anterior ou igual √† do outro."""
        return self.data_hora <= outro.data_hora

    def __gt__(self, outro):
        """Compara se a data e hora deste evento √© posterior √† do outro."""
        return self.data_hora > outro.data_hora

    def __ge__(self, outro):
        """Compara se a data e hora deste evento √© posterior ou igual √† do outro."""
        return self.data_hora >= outro.data_hora

# Testando a classe atualizada

# Criando duas inst√¢ncias de Evento com datas diferentes
evento1 = Evento(
    titulo="Reuni√£o de Planejamento",
    data_hora=datetime(2023, 10, 15, 14, 30),
    descricao="Reuni√£o para planejar o pr√≥ximo trimestre."
)

evento2 = Evento(
    titulo="Festa de Anivers√°rio",
    data_hora=datetime(2023, 12, 25, 20, 0),
    descricao="Festa de anivers√°rio do Jo√£o."
)

# Imprimindo as inst√¢ncias
print("Evento 1:")
print(evento1)
print("\nEvento 2:")
print(evento2)

# Testando os m√©todos de compara√ß√£o
print("\nCompara√ß√µes entre eventos:")
print(f"evento1 == evento2: {evento1 == evento2}")  # False
print(f"evento1 != evento2: {evento1 != evento2}")  # True
print(f"evento1 < evento2: {evento1 < evento2}")    # True
print(f"evento1 <= evento2: {evento1 <= evento2}")  # True
print(f"evento1 > evento2: {evento1 > evento2}")    # False
print(f"evento1 >= evento2: {evento1 >= evento2}")  # False
```
