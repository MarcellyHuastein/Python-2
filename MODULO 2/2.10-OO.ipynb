`EventoABC` também possui os atributos `_titulo` (string) e `_descricao` (string), cujos valores são recebidos e inicializados no construtor da classe. Note a convenção de nomenclatura indicando o caráter privado desses atributos.

````python
from datetime import datetime

class Evento(EventoABC):
    def __init__(self, titulo, descricao, data_hora):
        """
        Construtor da subclasse Evento.
        Inicializa os atributos da classe base e adiciona data_hora.
        """
        super().__init__(titulo, descricao)
        self.data_hora = data_hora
        self._is_concluido = False  # Atributo privado para armazenar o status de conclusão

    def __str__(self):
        """
        Implementação do método abstrato __str__.
        Retorna uma representação legível do evento.
        """
        return (f"Evento: {self._titulo}, "
                f"Data: {self.data_hora}, "
                f"Descrição: {self._descricao}, "
                f"Concluído: {self._is_concluido}")

    def isConcluido(self):
        """
        Implementação do método abstrato isConcluido.
        Verifica se a data e hora do evento já passaram.
        """
        if self.data_hora < datetime.now():
            self._is_concluido = True
        return self._is_concluido

# Testando a subclasse Evento
evento = Evento(
    titulo="Reunião de Planejamento",
    descricao="Reunião para planejar o próximo trimestre.",
    data_hora=datetime(2023, 10, 15, 14, 30)
)

print(evento)  # Usa o método __str__
print(f"Está concluído? {evento.isConcluido()}")  # Usa o método isConcluido
````

````python
from datetime import datetime, timedelta

class DataHora:
    # Atributo de classe para formatação de data e hora
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self):
        # Atributo privado para armazenar a data e hora
        self._data_hora = None

    @property
    def data_hora(self):
        """
        Getter da propriedade data_hora.
        Retorna a data e hora formatada como string.
        """
        if self._data_hora is None:
            return "Data e hora não definidas"
        return self._data_hora.strftime(self.FORMAT)

    @data_hora.setter
    def data_hora(self, value):
        """
        Setter da propriedade data_hora.
        Converte uma string formatada em datetime e armazena em _data_hora.
        Lança ValueError se a string for inválida.
        """
        try:
            self._data_hora = datetime.strptime(value, self.FORMAT)
        except ValueError as e:
            raise ValueError(f"Formato de data inválido. Use o formato: {self.FORMAT}") from e

    def isPassado(self):
        """
        Verifica se a data e hora armazenadas já passaram.
        Retorna True se a data for no passado, False caso contrário.
        """
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.now()

    def somaDias(self, num_dias):
        """
        Soma um número de dias à data e hora armazenadas.
        Retorna a nova data e hora formatada como string.
        """
        if self._data_hora is None:
            return "Data e hora não definidas"
        data_hora_somada = self._data_hora + timedelta(days=num_dias)
        return data_hora_somada.strftime(self.FORMAT)

# Testando a classe DataHora
if __name__ == "__main__":
    # Instanciando o objeto
    dh = DataHora()

    # Definindo a data_hora através da propriedade
    dh.data_hora = '05/02/2024, 12:30'

    # Editando a data_hora através da função somaDias
    dh.data_hora = dh.somaDias(30)

    # Imprimindo a data_hora editada e se é passado
    print(dh.data_hora, dh.isPassado())
````

#### Q3.

````python
from datetime import datetime
from abc import ABC, abstractmethod

# Classe DataHora (definida anteriormente)
class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self):
        self._data_hora = None

    @property
    def data_hora(self):
        if self._data_hora is None:
            return "Data e hora não definidas"
        return self._data_hora.strftime(self.FORMAT)

    @data_hora.setter
    def data_hora(self, value):
        try:
            self._data_hora = datetime.strptime(value, self.FORMAT)
        except ValueError as e:
            raise ValueError(f"Formato de data inválido. Use o formato: {self.FORMAT}") from e

    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.now()

    def somaDias(self, num_dias):
        if self._data_hora is None:
            return "Data e hora não definidas"
        data_hora_somada = self._data_hora + timedelta(days=num_dias)
        return data_hora_somada.strftime(self.FORMAT)

# Classe abstrata EventoABC (definida anteriormente)
class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def __str__(self):
        pass

    @abstractmethod
    def isConcluido(self):
        pass

# Classe EventoUnico
class EventoUnico(EventoABC):
    def __init__(self, titulo, descricao, data_hora_str):
        """
        Construtor da classe EventoUnico.
        Inicializa os atributos da superclasse e o atributo _data_hora.
        """
        super().__init__(titulo, descricao)
        self._data_hora = DataHora()  # Instância da classe DataHora
        self._data_hora.data_hora = data_hora_str  # Define a data e hora através da propriedade

    def isConcluido(self):
        """
        Implementação do método abstrato isConcluido.
        Verifica se a data e hora do evento já passaram.
        """
        return self._data_hora.isPassado()

    def __str__(self):
        """
        Implementação do método abstrato __str__.
        Retorna uma representação legível do evento.
        """
        return (f"Evento: {self._titulo}, "
                f"Data: {self._data_hora.data_hora}, "
                f"Descrição: {self._descricao}, "
                f"Concluído: {self.isConcluido()}")

    def editar_data_hora(self, nova_data_hora_str):
        """
        Método para editar a data e hora do evento.
        Recebe uma string formatada e atualiza _data_hora.
        """
        self._data_hora.data_hora = nova_data_hora_str

# Testando a classe EventoUnico
if __name__ == "__main__":
    # Criar evento
    evento = EventoUnico('Reunião', 'Sala 302, prédio da esquina', '05/10/2023, 16:30')
    print(evento)

    # Editar data do evento (através do método editar_data_hora)
    evento.editar_data_hora('05/10/2024, 16:30')
    print(evento)
````

* Crie o método `editar_data_hora` que recebe `data_hora_antiga` e `data_hora_nova` e altera o elemento da coleção de objetos `DataHora` que corresponde a `data_hora_antiga` fornecida.

````python
from datetime import datetime, timedelta

# Classe DataHora (definida anteriormente)
class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self):
        self._data_hora = None

    @property
    def data_hora(self):
        if self._data_hora is None:
            return "Data e hora não definidas"
        return self._data_hora.strftime(self.FORMAT)

    @data_hora.setter
    def data_hora(self, value):
        try:
            self._data_hora = datetime.strptime(value, self.FORMAT)
        except ValueError as e:
            raise ValueError(f"Formato de data inválido. Use o formato: {self.FORMAT}") from e

    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.now()

    def somaDias(self, num_dias):
        if self._data_hora is None:
            return "Data e hora não definidas"
        data_hora_somada = self._data_hora + timedelta(days=num_dias)
        return data_hora_somada.strftime(self.FORMAT)

# Classe abstrata EventoABC (definida anteriormente)
class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def __str__(self):
        pass

    @abstractmethod
    def isConcluido(self):
        pass

# Classe EventoRecorrente
class EventoRecorrente(EventoABC):
    def __init__(self, titulo, descricao, data_hora_inicial, data_hora_final, intervalo_repeticao):
        """
        Construtor da classe EventoRecorrente.
        Inicializa os atributos da superclasse e a lista de datas e horas.
        """
        super().__init__(titulo, descricao)
        self._datas_horas = []  # Lista de objetos DataHora
        self._preencher_datas_horas(data_hora_inicial, data_hora_final, intervalo_repeticao)

    def _preencher_datas_horas(self, data_hora_inicial, data_hora_final, intervalo_repeticao):
        """
        Preenche a lista de datas e horas com base no intervalo de repetição.
        """
        data_atual = DataHora()
        data_atual.data_hora = data_hora_inicial

        data_final = DataHora()
        data_final.data_hora = data_hora_final

        while datetime.strptime(data_atual.data_hora, DataHora.FORMAT) <= datetime.strptime(data_final.data_hora, DataHora.FORMAT):
            nova_data = DataHora()
            nova_data.data_hora = data_atual.data_hora
            self._datas_horas.append(nova_data)
            data_atual.data_hora = data_atual.somaDias(intervalo_repeticao)

    def isConcluido(self, indice):
        """
        Verifica se a ocorrência do evento no índice especificado já passou.
        """
        if indice < 0 or indice >= len(self._datas_horas):
            raise IndexError("Índice fora dos limites da lista de datas e horas.")
        return self._datas_horas[indice].isPassado()

    def __str__(self):
        """
        Retorna uma representação legível de todas as ocorrências do evento.
        """
        resultado = []
        for i, data_hora in enumerate(self._datas_horas):
            resultado.append(
                f"Evento: {self._titulo}, "
                f"Data: {data_hora.data_hora}, "
                f"Descrição: {self._descricao}, "
                f"Concluído: {self.isConcluido(i)}"
            )
        return "\n".join(resultado)

    def editar_data_hora(self, data_hora_antiga, data_hora_nova):
        """
        Edita uma data e hora específica na lista de ocorrências.
        """
        for data_hora in self._datas_horas:
            if data_hora.data_hora == data_hora_antiga:
                data_hora.data_hora = data_hora_nova
                return
        raise ValueError(f"Data e hora '{data_hora_antiga}' não encontrada na lista de ocorrências.")

# Testando a classe EventoRecorrente
if __name__ == "__main__":
    # Criar evento
    eventos = EventoRecorrente(
        'Reunião', 'Sala 302, prédio da esquina', 
        '05/01/2024, 16:30', '05/01/2025, 16:30', 30
    )

    # Imprimir eventos
    print(eventos)

    # Editar um dos eventos
    eventos.editar_data_hora('05/12/2024, 16:30', '05/12/2024, 11:30')

    # Imprimir eventos
    print("\nApós edição:")
    print(eventos)
````

````python
from datetime import datetime, timedelta

# Classe DataHora (definida anteriormente)
class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self):
        self._data_hora = None

    @property
    def data_hora(self):
        if self._data_hora is None:
            return "Data e hora não definidas"
        return self._data_hora.strftime(self.FORMAT)

    @data_hora.setter
    def data_hora(self, value):
        try:
            self._data_hora = datetime.strptime(value, self.FORMAT)
        except ValueError as e:
            raise ValueError(f"Formato de data inválido. Use o formato: {self.FORMAT}") from e

    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.now()

    def somaDias(self, num_dias):
        if self._data_hora is None:
            return "Data e hora não definidas"
        data_hora_somada = self._data_hora + timedelta(days=num_dias)
        return data_hora_somada.strftime(self.FORMAT)

# Classe abstrata EventoABC (definida anteriormente)
class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def __str__(self):
        pass

    @abstractmethod
    def isConcluido(self):
        pass

# Classe EventoUnico (definida anteriormente)
class EventoUnico(EventoABC):
    def __init__(self, titulo, descricao, data_hora_str):
        super().__init__(titulo, descricao)
        self._data_hora = DataHora()
        self._data_hora.data_hora = data_hora_str

    def isConcluido(self):
        return self._data_hora.isPassado()

    def __str__(self):
        return (f"Evento: {self._titulo}, "
                f"Data: {self._data_hora.data_hora}, "
                f"Descrição: {self._descricao}, "
                f"Concluído: {self.isConcluido()}")

    def editar_data_hora(self, nova_data_hora_str):
        self._data_hora.data_hora = nova_data_hora_str

# Classe EventoRecorrente (definida anteriormente)
class EventoRecorrente(EventoABC):
    def __init__(self, titulo, descricao, data_hora_inicial, data_hora_final, intervalo_repeticao):
        super().__init__(titulo, descricao)
        self._datas_horas = []
        self._preencher_datas_horas(data_hora_inicial, data_hora_final, intervalo_repeticao)

    def _preencher_datas_horas(self, data_hora_inicial, data_hora_final, intervalo_repeticao):
        data_atual = DataHora()
        data_atual.data_hora = data_hora_inicial

        data_final = DataHora()
        data_final.data_hora = data_hora_final

        while datetime.strptime(data_atual.data_hora, DataHora.FORMAT) <= datetime.strptime(data_final.data_hora, DataHora.FORMAT):
            nova_data = DataHora()
            nova_data.data_hora = data_atual.data_hora
            self._datas_horas.append(nova_data)
            data_atual.data_hora = data_atual.somaDias(intervalo_repeticao)

    def isConcluido(self, indice=0):
        if indice < 0 or indice >= len(self._datas_horas):
            raise IndexError("Índice fora dos limites da lista de datas e horas.")
        return self._datas_horas[indice].isPassado()

    def __str__(self):
        resultado = []
        for i, data_hora in enumerate(self._datas_horas):
            resultado.append(
                f"Evento: {self._titulo}, "
                f"Data: {data_hora.data_hora}, "
                f"Descrição: {self._descricao}, "
                f"Concluído: {self.isConcluido(i)}"
            )
        return "\n".join(resultado)

    def editar_data_hora(self, data_hora_antiga, data_hora_nova):
        for data_hora in self._datas_horas:
            if data_hora.data_hora == data_hora_antiga:
                data_hora.data_hora = data_hora_nova
                return
        raise ValueError(f"Data e hora '{data_hora_antiga}' não encontrada na lista de ocorrências.")

# Criando uma lista de eventos
lista_eventos = [
    EventoUnico('Reunião de Planejamento', 'Sala 302, prédio da esquina', '05/10/2023, 16:30'),
    EventoRecorrente('Reunião Mensal', 'Sala 302, prédio da esquina', '05/01/2024, 16:30', '05/01/2025, 16:30', 30),
    EventoUnico('Apresentação de Projeto', 'Auditório Principal', '15/11/2023, 10:00'),
    EventoRecorrente('Treinamento Semanal', 'Sala de Treinamento', '01/01/2024, 09:00', '31/12/2024, 09:00', 7)
]

# Iterando sobre a lista de eventos e imprimindo cada um
for evento in lista_eventos:
    print(evento)
    print("-" * 50)  # Separador para melhor visualização
````
